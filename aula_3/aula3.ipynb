VAMOS COMEÇAR A ANALISAR E OTIMIZAR OS DADOS

```python
print(f"Uso de memória ANTES da otimização (Real): {df.memory_usage(deep=True).sum() / 1024:.2f} KB")
print(f"Uso de memória ANTES da otimização (Padrão): {df.memory_usage(deep=False).sum() / 1024:.2f} KB")
```

dropar colunas que não são tão úteis para analisar

```python
df = df.drop(columns=["coffee_per_day", "breaks_per_day", "night_usage"])
df
```

```python
print(f"Uso de memória DEPOIS de remover as colunas (Real): {df.memory_usage(deep=True).sum() / 1024:.2f} KB")
print(f"Uso de memória DEPOIS de remover as colunas (Padrão): {df.memory_usage(deep=False).sum() / 1024:.2f} KB")
```

```python
df["addiction_level"] = df["addiction_level"].astype("category")
```

```python
print(f"Uso de memória DEPOIS de converter para 'category' (Real): {df.memory_usage(deep=True).sum() / 1024:.2f} KB")
print(f"Uso de memória DEPOIS de converter para 'category' (Padrão): {df.memory_usage(deep=False).sum() / 1024:.2f} KB")
```

```python
df.describe().T
```

```python
df.info()
```

```python
df["age"] = df["age"].astype("Int8")
```

```python
df["daily_screen_time"] = df["daily_screen_time"].astype("float32")
df["social_media_hours"] = df["social_media_hours"].astype("float32")
df["study_hours"] = df["study_hours"].astype("float32")
df["sleep_hours"] = df["sleep_hours"].astype("float32")
df["notifications_per_day"] = df["notifications_per_day"].astype("float32")
df["focus_score"] = df["focus_score"].astype("float32")
df["distraction_score"] = df["distraction_score"].astype("float32")
df["productivity_score"] = df["productivity_score"].astype("float32")

```

```python
print("Uso de memória FINAL (Real): {:.2f} KB".format(df.memory_usage(deep=True).sum() / 1024))
print("Uso de memória FINAL (Padrão): {:.2f} KB".format(df.memory_usage(deep=False).sum() / 1024))
```

agora nossos dados estao bem mais otimizados

```python
df.to_csv('student_phone_addiction_affects_on_productivity(in).csv', index=False) # as vezesn n salva tudo certinho
df.to_parquet('student_phone_addiction_affects_on_productivity(in).parquet', index=False) # mais eficiente
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) # Define um tamanho padrão para as figuras
```

```python
# Criando uma figura com dois subplots (um ao lado do outro)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histograma para a variável 'Age'
sns.histplot(df["age"], bins=11, kde=True, ax=axes[0])
axes[0].set_title('Distribuição da Idade')
axes[0].set_xlabel('Idade')
axes[0].set_ylabel('Frequência')

# Histograma para a variável daily_screen_time'
sns.histplot(df["daily_screen_time"], bins=10, kde=True, ax=axes[1], color='skyblue')
axes[1].set_title('Distribuição Diária de Tempo de Tela (horas)')
axes[1].set_xlabel('Tempo de Tela (horas)')
axes[1].set_ylabel('Frequência')

# Ajusta o layout para não haver sobreposição
plt.tight_layout()
plt.show()
```

usando box plot

```python
# Seleciona apenas as colunas numéricas de interesse
numerical_cols = ['age', 'daily_screen_time', 'study_hours', 'sleep_hours']

# Cria o box plot usando seaborn
sns.boxplot(data=df[numerical_cols])
plt.title('Distribuição e Escala das Variáveis Numéricas')
plt.ylabel('Valor')
plt.xlabel('Variáveis')
plt.show()
```

grafico em barra para coluna vicio

```python
# Criando uma figura com dois subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico de Barras para 'addiction_level'
sns.countplot(x='addiction_level', data=df, ax=axes[0], palette='viridis', order = df['addiction_level'].value_counts().index)
axes[0].set_title('Contagem por Nível de Dependência')
axes[0].set_xlabel('Nível de Dependência')
axes[0].set_ylabel('Contagem')

# Gráfico de Barras para 'age'
sns.countplot(x='age', data=df, ax=axes[1], palette='plasma')
axes[1].set_title('Contagem por Idade')
axes[1].set_xlabel('Idade')
axes[1].set_ylabel('Contagem')

plt.tight_layout()
plt.show()
```

comparar variaveis em boxplots

```python
sns.boxplot(x='age', y='daily_screen_time', data=df, palette='pastel')
plt.title('Tempo de Tela por Idade')
plt.xlabel('Idade')
plt.ylabel('Tempo de Tela (horas)')
plt.show()
```

```python
# comparar idade com produtividade

sns.boxplot(x='age', y='productivity_score', data=df, palette='pastel')
plt.title('Produtividade por Idade')
plt.xlabel('Idade')
plt.ylabel('Produtividade')
plt.show()
```
