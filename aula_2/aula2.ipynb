para saber os dados que foram escritos em uma coluna, sem repetição

```python
df["gender"].unique()
```

```python
df["known_for_department"].unique()
```

para tratar os dados e alterar o tipo da variavel

```python
df["gender"] = df["gender"].map({0: "Not Set", 1: "Female", 2: "Male", 3: "Non-Binary"}).astype("category")
```

```python
df["known_for_department"].map({"Acting": "Arts", "Directing": "Arts", "Production": "Arts", "Writing": "Arts", "Creator": "Arts", "Sound": "Technical", "Camera": "Technical", "Visual Effects": "Technical", "Editing": "Technical", "Art": "Technical", "Costume & Make-Up": "Technical", "Crew": "Technical", "Lighting": "Technical"}).fillna("Other").astype("category")
```

```python
df["name"] = df["name"].astype("string[pyarrow]")
```

```python
df["popularity"] = df["popularity"].astype("float32")
```

para descrever a qntd de dados, media, valor min, valor max em cada coluna

```python
df.describe(include="all")
```
